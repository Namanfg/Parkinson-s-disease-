Import the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score,f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier,VotingClassifier
from sklearn.linear_model import LogisticRegression,Perceptron
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
```

Data Preprocessing

```python
x=df.drop(columns=['name','status'],axis=1)
y=df['status']
```

```python
x
```

```python
print(y)
```

Training and test data split


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(f"The shape of the training set for x data is {x_train.shape} and the shape of the testing set for the x data is {x_test.shape}")
```

```python
print(f"The shape of the training set for y data is {y_train.shape} and the shape of the testing set for the y data is {y_test.shape}")
```

Data Standardization

```python
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

Model Training



Support Vector Machine

```python
# SVC --- Support Vector Classifier
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

KNN

```python
#KNN -- k-nearest neighbour
model_1 = KNeighborsClassifier(n_neighbors=14)
model_1.fit(x_train,y_train)
```

RANDOM FOREST CLASSIFIER

```python
#Random Forest Classifier
model_2 = RandomForestClassifier(n_estimators=5,criterion="entropy")
model_2.fit(x_train,y_train)
```

NAIVE BAYES

```python
#Guassian Naive Bayes
model_3 = GaussianNB()
model_3.fit(x_train,y_train)
```

LOGISTIC REGRESSION

```python
model_4 = LogisticRegression()
model_4.fit(x_train,y_train)
```

PERCEPTRON

```python
model_5 = Perceptron()
model_5.fit(x_train,y_train)
```

Model Evaluation

EVALUATION FOR SVM

```python
# accuracy score on training data
x_train_pred = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train,x_train_pred)
```

```python
training_data_accuracy
```

```python
# accuracy score on testing data
x_test_pred = model.predict(x_test)
testing_data_accuracy = accuracy_score(y_test,x_test_pred)
```

XG BOOST

```python
model_xgb = xgb.XGBClassifier(booster='dart',max_depth=16,learning_rate=0.4,alpha=3.2)
model_xgb.fit(x_train,y_train)
```

SOFT VOTING

```python
voting_classifier_soft = VotingClassifier(
    estimators = [('knn_model',KNeighborsClassifier(n_neighbors=14)),
                  ('lr_model',LogisticRegression()),
                  ('SVC_model',svm.SVC(gamma='auto',probability=True)),
                  ],
    voting='soft'
)
voting_classifier_soft.fit(x_train,y_train)
```

```python
x_train_pred_6 = voting_classifier_soft.predict(x_train)
training_data_accuracy_vcs = accuracy_score(y_train,x_train_pred_6)
training_data_accuracy_vcs
```

```python
x_test_pred_6 = voting_classifier_soft.predict(x_test)
test_data_accuracy_vcs = accuracy_score(y_test,x_test_pred_6)
test_data_accuracy_vcs
```

HARD VOTING

```python
voting_classifier_hard = VotingClassifier(
    estimators = [('knn_model',KNeighborsClassifier(n_neighbors=14)),
                  ('lr_model',LogisticRegression()),
                  ('SVC_model',svm.SVC(gamma='auto',probability=True)),
                  ],
    voting='hard'
)
voting_classifier_hard.fit(x_train,y_train)
```

```python
x_train_pred_7 = voting_classifier_hard.predict(x_train)
training_data_accuracy_vch = accuracy_score(y_train,x_train_pred_7)
training_data_accuracy_vch
```

```python
x_test_pred_7 = voting_classifier_hard.predict(x_test)
test_data_accuracy_vch = accuracy_score(y_test,x_test_pred_7)
test_data_accuracy_vch
```

Since the training data and the testing data accuracy score are close so that means the model has not been overfitted or underfitted
